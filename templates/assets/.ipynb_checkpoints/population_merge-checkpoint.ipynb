```python
regional_df = pd.read_csv("cont_data.csv")
pop_df = pd.read_csv("pop_csv.csv")
```

```python
regional_df = regional_df.rename(columns={"Three_Letter_Country_Code": "Country Code"})
```

```python
pop_df = pop_df.rename(columns={"Country Name": "Country"})
```

```python
merged_df = pd.merge(esg_df, pop_df, on="Country")
```

```python
final_merged_df = merged_df.merge(regional_df, on='Country Code', how = 'left')
```

```python
final_merged_df
```

```python
merged_df.to_csv("merged_csv.csv",index=False)
```
